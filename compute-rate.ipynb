```python
import pandas as pd

population_16 = pd.read_csv('pop-by-zip-code.csv')
complaints = pd.read_csv('total-complaints-zipcode.csv')
```

```python
for index, row in complaints.iterrows():
    zipcode = row['ZIPCODE']
    complaint_count = row['COMPLAINT_COUNT']
    pop_16_exist = population_16[population_16["zip_code"] == zipcode]["aggregate"].values
    if len(pop_16_exist) != 0 and pop_16_exist[0] != 0 and pop_16_exist[0] >= 100:
        pop_16 = pop_16_exist[0]
        print(zipcode, round(complaint_count/pop_16, 2))
```
